Importing the dependencies


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Data collection and analysis

df = pd.read_csv('./content/diabetes.csv')
```

```python
# Seprating the data and labels 
```

```python
x = df.drop(columns=['Outcome'], axis =1)
y = df['Outcome']
```

```python
x
```

```python
# data Standarization

scalar = StandardScaler() # create instance of standardscalar

scalar.fit(x)  # fit the data

standarized_data = scalar.transform(x) # transform it into standarized values
```

```python
x = standarized_data
```

Train  and Train Splitting

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 4, stratify = y) # stratify ->divide the date in unformity
```

```python
print(x_train.shape, x_test.shape)
```

Model training

```python
classifier  = svm.SVC(kernel='linear') # SVC - support vector classifier
```

```python
# training the model

classifier.fit( x_train, y_train)
```

Model Evaluation

```python
#Finding the Accuracy Score
# Accuracy score on the training data

x_train_predict = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predict, y_train)

print(training_data_accuracy)
```

```python
#Finding the Accuracy Score
# Accuracy score on the testing data

x_test_predict = classifier.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_predict, y_test)

print(testing_data_accuracy)
```

making an predictive system

```python
# 4, 110, 92, 0 ,0, 37.6, 0.191, 30, 0

input_data = (4,110,92,0,0,37.6,0.191,30)
#changeing the input data into numpy array

input_data_npArray= np.asarray(input_data)

print(input_data_npArray)

#Reshape the array as we are predicting the for one instances
input_data_reshaped = input_data_npArray.reshape(1, -1)

print(input_data_reshaped)

# Standardize the input data

std_data  = scalar.transform(input_data_reshaped)
print(std_data)

#Make the prediction finally 

prediction = classifier.predict(std_data)
print(prediction)
```
